```
# 코모란(Komoran) 토큰화
from konlpy.tag import Komoran
komoran= Komoran()
komoran_tokens = komoran.morphs(docs)
print(komoran_tokens)
len(komoran_tokens)
```

```
import pandas as pd

uni_token = pd.unique(komoran_tokens)
len(uni_token)
```

```
# 문서 내 토큰등장빈도 수 딕셔너리 생성
from collections import defaultdict

key_word = {}
for i, u in enumerate(uni_token):
    count = 0
    for j, t in enumerate(komoran_tokens):
        if u == t:
            count += 1
    key_word.setdefault(u,count)
key_word
```

```
# 중요단어 결정
imp_word = []
for i, (key, elem) in enumerate(key_word.items()): # 딕셔너리 인덱스 접근
    if 0.001 < elem/len(komoran_tokens) < 0.2:
        imp_word.append(key)
imp_word
```

```
# 문서를 문장으로 토큰화
docs_tokens = docs.split('.')
docs_tokens.pop() # 마지막 공백 제거
docs_tokens
```

```
# 문장중요도 계산
sent_sig = {}
for i, v in enumerate(docs_tokens):
    komoran_tokens2 = komoran.morphs(v)
    # print(komoran_tokens2)
    window = {}
    for idx, k in enumerate(komoran_tokens2):
        if k in imp_word:
             window[k] = idx
    # print(window)

    window_start = min(window.values())
    window_end = max(window.values())
    len_word = window_end-window_start+1 # 윈도 내 포함된 단어 개수
    num_import = len(window) # 윈도 내 포함된 중요단어 개수

    # 문장 중요도
    sig_f = num_import**2/len_word
    sent_sig[v] = sig_f
```
