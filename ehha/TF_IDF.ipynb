## TF-IDF

```python
import numpy as np
import pandas as pd
```

```python
d1 = "The cat sat on my face I hate a cat"
d2 = "The dog sat on my bed I love a dog"

text_list = [d1, d2]

vocab = list(set(w for doc in text_list for w in doc.split()))
vocab.sort()
vocab
```

```python
N = len(text_list)        
N
```

```python
# TF, IDF, TF-IDF 구하는 함수
def f(t, d):
    return d.count(t)

def tf(t, d):
    return f(t,d)/len(d)

def idf(t):
    df = 0
    for doc in text_list:
        df += t in doc
    return np.log(N/df)

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
# TF(DTM)
result = []
for i in range(N):
    result.append([])
    d = text_list[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t, d))
        
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
# IDF
result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))
    
idf_ = pd.DataFrame(result, index = vocab, columns = ["IDF"])
idf_
```

```python
# TF-IDF
result = []
for i in range(N):
    result.append([])
    d = text_list[i]
    for j in range(len(vocab)):
        t = vocab[j]
        
        result[-1].append(tfidf(t,d))
tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```
