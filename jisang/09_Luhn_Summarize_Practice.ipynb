```
import docx2txt

# 문서 경로 지정
import_docs = docx2txt.process("/content/docs.docx")

# 문서 출력
import_docs
```

```
# split을 이용한 문서 전처리
docs_split = import_docs.split('.')

docs = []
for line in docs_split:
    docs.append(line.replace("\n\n", ""))

docs
```

```
# tokenize를 이용한 데이터 전처리
from nltk.tokenize import sent_tokenize

import nltk
nltk.download('punkt')

docs = sent_tokenize(import_docs)

docs_line = []

for line in docs:
    if '\n\n' in line:
        line_split = line.split('\n\n')
        docs_line.append(line_split[:])
    else:
        docs_line.append(line)

docs_line
```

## **2-2. Luhn Summarize 구현**

```
from konlpy.tag import Mecab

import numpy as np
import math
import networkx as nx

#mecab = Mecab()

# 토큰화 : konlpy를 이용한 토큰화 방안 - 코로나19를 인식하지 못함
# for line in docs:
#     tokens.append([token for token in mecab.pos(line)])

# 토큰화 : 어절 별로 토큰화 방안
tokens = []
for line in docs:
    tokens.append(line.split())

tokens[0:3]
```

```
# 전체 어절 개수
total_count = 0
words = []

for line in tokens:
    for word in line:
        total_count += 1
        words.append(word)

print(total_count, words)
```

```
# 어절별 빈도수 확인
import nltk

freq = nltk.FreqDist(words)

freq
```

```
# 어절별 비율 확인
word_rate = {}

for word in words:
    word_rate[word] =freq[word]/total_count

word_rate
```

```
# 중요단어 결정
important_word = {}

for word in words:
    if word_rate[word] < 0.05 and word_rate[word] > 0.01:
        important_word[word] = word_rate[word]

print("=========== 핵심키워드 ==========")
for w in important_word:
    print(w, ":" ,important_word[w])
```

```
# 문장별 중요도 계산
line_important = []

for line in tokens:
    loc_important = []
    important_count = 0
    for loc, words in enumerate(line):
        for im_word in important_word:
            if im_word in words:
                loc_important.append(loc)
                important_count += 1
    try:
        line_important.append((loc_important[-1] - loc_important[0] + 1)**2/important_count)
    except:
        line_important.append(0)

line_important
```

```
# 중요한 문장 출력
important_line = {}

for num, line in enumerate(docs):
    important_line[line] = line_important[num]

sorted(important_line.items(), reverse=True, key=lambda item:item [1])
```
