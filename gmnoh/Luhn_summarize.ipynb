```python
# Kkma 토큰화
from konlpy.tag import Kkma
kkma= Kkma()
kkma_tokens = kkma.morphs(doc)

print(kkma_tokens)
print(len(kkma_tokens))
```

```python
# Kkma 품사태깅
kkmaTag = []
for token in kkma_tokens:
    kkmaTag += kkma.pos(token)
    
print(kkmaTag)
```

```python
# NNG만 고려한 Kkma 토큰화
kkma_tokens = []

for i in kkmaTag:
    if i[1] == 'NNG':
        kkma_tokens.append(i[0])

print(kkma_tokens)
print(len(kkma_tokens))
```

```python
# 문서내 단어 빈도
keys = list(set(kkma_tokens))
zero = [0 for i in range(len(keys))]
freq = dict(zip(keys, zero))

for i in keys:
    freq[i] = kkma_tokens.count(i)

print(freq)
```

```python
# 중요단어 결정
word = []
word_count = len(kkma_tokens)

def get_word(min, max):
    for i in freq:
        if freq[i] / word_count > min and freq[i] / word_count < max:
            word.append(i)
```

```python
get_word(0.01, 0.5)

print(word)
```

```python
# 문장단위로 나누기
doc_split = []

doc_split_ = doc.split('.')
doc_split_.pop()

for i in doc_split_:
    tmp = i.split('\n')
    for j in tmp:
        if j != '' and j != ' ':
            doc_split.append(j)

for i in doc_split:
    print(i)
    print('----------------------------------------')
```

```python
# 문장 중요도 계산
importance = []

for i in doc_split:
    sig_word = []
    tokens = Kkma().morphs(i)

    for j in tokens:
        if j in word:
            sig_word.append(tokens.index(j))

    length = max(sig_word) - min(sig_word) + 1
    importance.append(len(sig_word)*len(sig_word) / length)

print(importance)
```

```python
# 문장 중요도 순위별 출력
importance_cpy = importance.copy()
importance_cpy.sort(reverse=True)
importance_cpy

for i in importance_cpy:
    print(doc_split[importance.index(i)])
    print('----------------------------')
```
